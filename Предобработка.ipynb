```python
from natasha import MorphVocab
```

```python
#файл test
add=[]
for _ in range(10000):
  text=tb['text'][_]
  doc = Doc(text)
  doc.segment(segmenter)
  doc.tag_morph(morph_tagger)
  doc.parse_syntax(syntax_parser)
  morph_vocab = MorphVocab()

  for token in doc.tokens:
    token.lemmatize(morph_vocab)

  s = { _.lemma for _ in doc.tokens}
  p=[word for word in s if word not in stopWords]
  add.append(p)
df1=pd.DataFrame({'text':add})
df1.to_csv('data10t.csv')
!cp data10t.csv "drive/My Drive/"
```

```python
#файл  val
add=[]
for _ in range(10000):
  text=tb2['text'][_]
  doc = Doc(text)
  doc.segment(segmenter)
  doc.tag_morph(morph_tagger)
  doc.parse_syntax(syntax_parser)
  morph_vocab = MorphVocab()

  for token in doc.tokens:
    token.lemmatize(morph_vocab)

  s = { _.lemma for _ in doc.tokens}
  p=[word for word in s if word not in stopWords]
  add.append(p)
df2=pd.DataFrame({'text':add})
df2.to_csv('data10v.csv')
!cp data10v.csv "drive/My Drive/"
```
